#### Age vs spending
    1. Are younger customers spending less or buying specific categories?
    2. Do older customers respond better to campaigns?

```python
df['Age'].sort_values(ascending = False)
# min = 24
# max = 80
```

```python
Age_Range = [(21,30),(31,40),(41,50),(51,60),(61,70),(71,100)]

def define_age_range(Age):
    for i in Age_Range:
        if i[0] <= Age <= i[1]:
            return f'{i[0]} - {i[1]}'
    return 'unknown'

        
df['Age_Range'] = df['Age'].apply(define_age_range)
```

```python
df.head()
```

```python
df[['Age', 'Age_Range']]
```

```python
amount_age = pd.DataFrame(df.groupby(by = 'Age_Range')['MntTotal'].mean())
amount_age
```

```python
amount_age['MntTotal'].plot.barh(title = "Average amount spent per age range", ylabel ="Age Range", figsize =(9,10))

# 71-100 seems to spend the most per person
# 21-30 is the second place
```

```python
age_spend = (df.groupby('Age_Range')['MntTotal'].sum()).reset_index()
age_spend
```

```python
sns.barplot(data = age_spend, x = 'Age_Range', y='MntTotal')
plt.title('Amount spent per Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total amount spent')
plt.show()
```

```python
age_count = (df['Age_Range'].value_counts()).reset_index()
age_count = age_count.rename(columns = {'index':'Age_Range', 'Age_Range':'Count'})
age_count
# 41-50: largest group
# 21-30: samllest group
```

```python
accepted_cmp = df[df['AcceptedCmpOverall'] != 0]
accepted_cmp
```

```python
accepted_counts = accepted_cmp['Age_Range'].value_counts()
accepted_counts = accepted_counts.reset_index()
accepted_counts = accepted_counts.rename(columns = {'index':'Age_Range', 'Age_Range':'Accepted_Count'})
# 41-50 is the largest age group
# 21-30 is the smallest age group
accepted_counts
```

```python
age_merged = pd.merge(age_count, accepted_counts, on = 'Age_Range')
age_merged
```

```python
age_merged['Accepted_Percentage']= age_merged['Accepted_Count'] / age_merged['Count']
age_merged = age_merged.sort_values(by = 'Age_Range')
age_merged
```

```python
sns.barplot(data = age_merged, x = 'Age_Range', y = 'Accepted_Percentage')
plt.title('Percentage of accepted campaign per age range')
# 21-30 and 71-100 accepts campagn at higher rate than all other 
```

```python
accepted_counts['Percentage'] = (accepted_counts['Accepted_Count'] / accepted_cmp.shape[0])
accepted_counts
```

```python
sns.barplot(data = accepted_counts, x = 'Age_Range', y='Percentage')
plt.title('Percentage of Accepted Campagn per Age Group')
plt.xlabel('Age Range')
plt.show()
```
